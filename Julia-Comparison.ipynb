# Julia Comparison

In this notebook we compare the output of the dynamical systems jl with our own code for a small configuration of the two layer L96.

```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import os
from tqdm.notebook import tqdm

from ginelli_plot import *
```

```python
directory = 'ginelli/step5'
ds = xr.open_mfdataset(directory + '/*.nc', parallel=True, combine='by_coords', concat_dim='time')
    
julia_data = xr.open_dataset('lyap55.jld2')['lambda'].values
julia_data = np.flip(np.sort(julia_data))
```

```python
cles = spectra(ds)
bles = spectra(ds, geometry ='B').values
```

```python
plt.plot(cles, '.-')
plt.plot(bles, '.-')
plt.plot(julia_data, '.-')
plt.grid()
plt.show()
```
